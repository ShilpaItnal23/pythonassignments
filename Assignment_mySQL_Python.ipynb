# 4 . Dump all the data into car table 

```python
import csv
with open('car.txt','r') as f:
    data = csv.reader(f, delimiter = '\n')
    for row in data:
        lst = row[0].split(',')
        print("INSERT INTO cardataset.car values('{0}','{1}','{2}','{3}','{4}','{5}','{6}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6]))
        cursor.execute( "INSERT INTO cardataset.car values('{0}','{1}','{2}','{3}','{4}','{5}','{6}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6]))
    mydb.commit()
```
