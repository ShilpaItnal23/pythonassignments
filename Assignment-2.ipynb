### 10. How can you tell the difference between break and continue?

### Both break and continue are used to control the loop.
### Both behave exactly opposite.
### In break statement when the condition becomes true it comes of the block and resumes the next statment.
### In continue statement, when the condition becomes true it does not comes out the block, instead it skips at ### that condition and resumes the loop.

```python
count = 0
i = 10
while(i>0):
    count = count + 1
    print(i)
    i =i-1
    if(i==5):
        break
```

```python
count = 0
i = 10
while(i>0):
    count = count + 1
    if(i==5):
        continue
    print(i)
    i =i-1
```

### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
for x in range(6):
  print(x) 
```

```python
for x in range(0,10):
  print(x) 
```

```python
for x in range(0,10,1):
  print(x)
```

### Syntax:
### range(start, stop, step)
### start	Optional. An integer number specifying at which position to start. Default is 0
### stop	Required. An integer number specifying at which position to stop (not included).
### step	Optional. An integer number specifying the incrementation. Default is 1
