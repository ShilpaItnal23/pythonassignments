```python
import csv
count = 0
with open('car.txt','r') as f:
    data = csv.reader(f, delimiter = '\n')
    for row in data:
        lst = row[0].split(',')
        print("INSERT INTO cardataset1.car1 values('{0}','{1}','{2}','{3}','{4}','{5}','{6}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6]))
        cursor.execute( "INSERT INTO cardataset1.car1 values('{0}','{1}','{2}','{3}','{4}','{5}','{6}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6]))
        count = count +1
    print(count)
mydb.commit()
```
